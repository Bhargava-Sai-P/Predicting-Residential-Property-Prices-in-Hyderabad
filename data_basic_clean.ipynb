```python
def clean(x):
    for i in range(0,len(x['price(L)'])):
        a=x.loc[i,'rate persqft'] 
        b=x.loc[i,'area insqft']
        m=np.round(float((a*b)/100000),2)
        n=x.loc[i,'price(L)']
        if m==n:
            continue
        else:
            x.loc[i,'price(L)']=m 
    #x=x.drop(labels='Unnamed: 0',axis=1)
    x=x.drop_duplicates()
    x.replace('not found', np.nan,inplace=True)
    x.replace('Not found', np.nan,inplace=True)
    x = x.astype({"latitude": float, "longitude": float,"locaLity score": float,"carpet area": float})
    writer = pd.ExcelWriter('data4.xlsx',engine ='xlsxwriter')
    x.to_excel(writer, 'Sheet1')
    writer.save()
```
